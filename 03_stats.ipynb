## Statsmodels

Rückbesinnung auf Theorie:
<br>Koeffizienten sind signifikant, wenn ihr p-Wert niedrig ist zb unter 0.01; deutet darauf hin, dass es eine *statistisch* bedeutsame Beziehung zwischen dieser unabhängigen Variablen und der abhängigen Variablen gibt.

<br>In der Zusammenfassung eines Regressionsmodells zeigt der Wert unter "coef" (Koeffizient) die geschätzte Veränderung der abhängigen Variablen für eine Einheit Veränderung der unabhängigen Variablen an, vorausgesetzt, alle anderen Variablen bleiben konstant.

```python
#es soll income durch Alter, Wochenstunden, Abschluss erklärt werden
model1 = smf.logit("income ~ age + hours_per_week + education_num", data=train_df)
```

```python
results1=model1.fit()
```

```python
print(results1.summary())
```

Pseudo R nur bei 0.19; alle Koeffizienten sind hochsignifikant; education_num hat den größten Einfluss auf income mit einem Koeffizienten von 0.345

```python
#es soll income durch Alter, Wochenstunden, Abschluss, Kapitalverlusten/-gewinnen erklärt werden
model2 = smf.logit("income ~ age + hours_per_week + education_num + capital_gain + capital_loss", data=train_df)
```

```python
results2=model2.fit()
```

```python
print(results2.summary())
```

Pseudo R etwas höher mit 0.26; alle Koeffizienten sind hochsiginifikant, wobei aber der Einfluss von capital_gain und capital_loss sehr gering ist.

```python
#es soll income durch Alter, Wochenstunden, Abschluss, Geschlecht und Familienstand erklärt werden
model3 = smf.logit("income ~ age + hours_per_week + education_num + sex + marital_status", data=train_df)
```

```python
results3=model3.fit()
```

```python
print(results3.summary())
```

In dieser Kombination wird der Koeffizient zu education_num sogar etwas größer, während die Koeffizienten age und hours_per_week an Erklärungskraft verlieren; sex male ist signifikant und hat einen eher geringen Einfluss (aber höher als die Kapital-Koeffizienten). Von den Familienstand-Kategorien sind signifikant nur die beiden married (AF und civ) und nerver-married. die married Kategorien haben einen im Vergleich hohen positiven Koeffizienten von 2,49 bzw 2,0; never-married hat einen negativen Einfluss (-0.529) auf income. pseudo R erneut gesteigert auf 0.328

```python
#es soll income durch Alter, wochenstunden, Abschluss, Geschlecht, Familienstand und Beruf erklärt werden
model4 = smf.logit("income ~ age + hours_per_week + education_num + sex + marital_status + occupation", data=train_df)
```

```python
results4=model4.fit()
```

```python
print(results4.summary())
```

Pseudo R konnte nochmals geringfügig gesteigert werden, nun 0.352; wobei die Koeffizienten hours_per_week und education_num etwas an Erklärungskraft einbüßen, während die anderen Koeffizienten zunehmen. Von den Berufen sind craft-repair und protect-priv-serv nicht signigikant; positiven Einfluss haben exec-managerial (Führungskräfte, 0.74), prof-specialty (Fachkräfte, 0.47), sales (Verkauf, 0.2 ) und tech-support (technischer Support, 0.56), die anderen Berufe haben einen  negativen Einfluss auf income, wobei farming-fishing den höchsten negativen Wert hat (-1.29) und other-services mit -0.84 ebenfalls einen recht hohen negativen Einfluss hat.

```python
#es soll income durch Alter, Wochenstunden, Abschluss, Geschlecht, Familienstand, Beruf und Herkunftsregion erklärt werden
model5 = smf.logit("income ~ age + hours_per_week + education_num + sex + marital_status + occupation + native_region", data=train_df)
```

```python
results5=model5.fit()
```

```python
print(results5.summary())
```

nur geringe Steigerung von pseudo R auf 0.353; die zusätzliche Variable native_region ist aber in den einzelnen Kategorien nicht signifikant, außer Mittelamerika geringfügig mit einem negativen Einfluss von -0.435 . Überlege diese Variable dann rauszulassen, da zusätzlich Verteilung sehr ungleich mit 90% "united-states"

```python
#es soll income durch Alter, wochenstunden, Abschluss, Geschlecht, Familienstand, Beruf und Kapitalverluste/-gewinne erklärt werden
model6 = smf.logit("income ~ age + hours_per_week + education_num + marital_status + sex + occupation + capital_gain + capital_loss", data=train_df)
```

```python
results6 = model6.fit()
```

```python
print(results6.summary())
```

hier wurden doch noch einmal die Kapital-Variablen dazu genommen - pseudo R wird zwar nochmals etwas höher (0.411) und die Koeffizienten sind signifikant, aber sie haben quasi keinen Einfluss auf die Veränderung der abhängigen Variablen, da die Koeffizienten nur 0.0003 bzw 0.0007 sind.

Zusammenfassend lässt sich also sagen, dass die Koeffizienten age, hours_per_week,  education_num, marital_status (married-AF, married-civ und never-married), sex (male), occupation (positiv: exec-managerial, prof-specialty, tech-support // negativ vor allem farming-fishing, other-service, handlers-cleaners) sowie capital_gain und capital_loss jeweils signifikant sind und Einfluss auf die abhängige Variable income haben, wobei die Berufe und der Familienstand die größten Veränderungen beim Einkommen hervorrufen.
