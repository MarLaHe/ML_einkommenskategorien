## Import

```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
PATH_TRAIN = os.path.join ("input", "adult.data")
```

```python
PATH_TEST = os.path.join ("input", "adult.test")
```

```python
TEMP_PATH = os.path.join("temp")
```

```python
train_df = pd.read_csv(os.path.join(PATH_TRAIN), header= None)
#spaltennamen
columns = ["age", 
           "workclass", 
           "final_weight", 
           "education", 
           "education_num", 
           "marital_status",
          "occupation",
          "relationship",
          "race",
          "sex",
          "capital_gain",
          "capital_loss",
          "hours_per_week",
           "native_country",
           "income"]
train_df.columns = columns
#leerzeichen raus und Ã¼berall kleinschreibung
for col in train_df.select_dtypes(include=['object']).columns:
    train_df[col] = train_df[col].str.strip().str.lower()
```

```python
train_df.sample(20)
```

```python
test_df = pd.read_csv(PATH_TEST, header= None)
columns = ["age", 
           "workclass", 
           "final_weight", 
           "education", 
           "education_num", 
           "marital_status",
          "occupation",
          "relationship",
          "race",
          "sex",
          "capital_gain",
          "capital_loss",
          "hours_per_week",
           "native_country",
           "income"]
test_df.columns = columns
for col in test_df.select_dtypes(include=['object']).columns:
    test_df[col] = test_df[col].str.strip().str.lower()
```

```python
train_df.to_pickle(os.path.join(TEMP_PATH, "train.pickle"))
```

```python
test_df.to_pickle(os.path.join(TEMP_PATH, "test.pickle"))
```
